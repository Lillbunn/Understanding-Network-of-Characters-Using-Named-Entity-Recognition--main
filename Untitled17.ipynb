```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_characters_in_the_Mahabharata"
page = requests.get(url)
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print (soup.prettify())
```

```python
scrapped_content = soup.findAll(attrs = {'class' : 'mw-headline'})
scrapped_content
```

```python
scrapped_content[1].text
```

```python
scrapped_content_1 = soup.findAll("p")
scrapped_content_1
```

```python
scrapped_content_1[2].text
```

```python
import csv
from bs4 import BeautifulSoup
# scrapped_content = soup.findAll(attrs={'class': 'mw-headline'})
# scrapped_content_1 = soup.findAll("p")

# Create a list of tuples combining the content from scrapped_content and scrapped_content_1
combined_data = list(zip(scrapped_content))

# Write the combined data into a CSV file
with open("combined_data_1.csv", "w", newline="", encoding="utf-8") as csvfile:
    csvwriter = csv.writer(csvfile)
    
    # Write the headers
    csvwriter.writerow(["Characters"])
    
    # Write the combined data row by row
    for row in combined_data:
        csvwriter.writerow([row[0].text.strip() if row[0] else "",])
```
